### Document Length Analysis
We analyze the distribution of summaries length to understand content variation.


```python
import numpy as np
import matplotlib.pyplot as plt

# Compute document lengths
df["summary_length"] = df["summaries"].apply(lambda x: len(x.split()))

# Display summary statistics
print(df["summary_length"].describe())

# Plot histogram of document lengths
plt.figure(figsize=(8,5))
plt.hist(df["summary_length"], bins=20, edgecolor='black')
plt.xlabel("Summary Length (words)")
plt.ylabel("Frequency")
plt.title("Distribution of Summary Lengths")
plt.show()
```

```python
import torch
from sentence_transformers import SentenceTransformer
from transformers import T5Tokenizer, T5ForConditionalGeneration
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import ast
import numpy as np

# Load dataset (assuming df is already loaded)
df["terms"] = df["terms"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Initialize embedding model
embedding_model = SentenceTransformer("BAAI/bge-small-en")

# Compute embeddings for dataset documents
document_texts = df["summaries"].tolist()  # Replace 'document_column' with actual column name
document_embeddings = embedding_model.encode(document_texts, convert_to_tensor=True)

# Function to retrieve top-k similar documents
def retrieve_documents(query, k=3):
    query_embedding = embedding_model.encode(query, convert_to_tensor=True)
    similarities = cosine_similarity(query_embedding.unsqueeze(0), document_embeddings)[0]  # Compute similarity
    top_k_indices = np.argsort(similarities)[-k:][::-1]  # Get top-k indices
    return df.iloc[top_k_indices]  # Return the top-k matching documents

# Initialize LLM (FLAN-T5)
t5_model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-small")
t5_tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-small")

# Function to generate response based on retrieved documents
def generate_response(query, k=3):
    retrieved_docs = retrieve_documents(query, k)
    context = " ".join(retrieved_docs["document_column"].tolist())  # Combine retrieved docs into context

    # Format prompt for T5
    prompt = f"Answer the following question based on the context:\n\nContext: {context}\n\nQuestion: {query}\n\nAnswer:"
    input_ids = t5_tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True).input_ids

    # Generate response
    output_ids = t5_model.generate(input_ids, max_length=150)
    answer = t5_tokenizer.decode(output_ids[0], skip_special_tokens=True)

    return answer

# Example query test
query = "What is the role of convolutional neural networks in computer vision?"
response = generate_response(query, k=3)
print("Generated Response:", response)
```
